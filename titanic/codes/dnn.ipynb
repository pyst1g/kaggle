```python
def preproceed(df,dummy_na=False):
    new_df = df.copy()
    # 欠損値の処理
    new_df.replace({'Nan':pd.np.nan,'nan':pd.np.nan},inplace=True)
    new_df.fillna(new_df.mean(numeric_only=True),inplace=True)
#     new_df['Fare'].fillna(new_df['Fare'].mean(),inplace=True)
#     new_df['Embarked'].fillna(new_df['Embarked'].mode().loc[0],inplace=True)
    
    #自分以外の家族の人数及び有無を追加
    new_df['NumFamily'] = new_df['SibSp'] + new_df['Parch']
    new_df['IsFamily'] = new_df['NumFamily'] != 0
    new_df = pd.concat([new_df, pd.get_dummies(new_df['Pclass'],prefix='IsPclass',dummy_na=dummy_na)],axis=1)
    new_df = pd.concat([new_df, pd.get_dummies(new_df['Embarked'],prefix='IsEmbarked',dummy_na=dummy_na)],axis=1)
    #質的変数をダミー変数に
#     new_df['IsPclass1'] = new_df['Pclass'] == 1
#     new_df['IsPclass2'] = new_df['Pclass'] == 2
#     new_df['IsPclass2'] = new_df['Pclass'] == 3
#     new_df['IsEmbarkedS'] = new_df['Embarked'] == 'S'
#     new_df['IsEmbarkedC'] = new_df['Embarked'] == 'C'
#     new_df['IsEmbarkedQ'] = new_df['Embarked'] == 'Q'
    
    
    #質的変数を変換'
    new_df['Sex'] = new_df['Sex'].apply(lambda x: 1 if x == 'male' else 0)
#     new_df['Embarked'] = new_df['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2}).astype('int')


    #いらない列(Cabin,Name,PassengerId,Ticket)の削除
    new_df.drop(['Cabin','Name','Ticket','Embarked'],axis=1,inplace=True)
    
    return new_df
```

```python
#csvデータの呼び出し場所
loadpath = "../input/"
#csvデータの保存場所
savepath = "../output/"
```

```python
#データ呼び出し
train = pd.read_csv(loadpath + "train.csv")
test = pd.read_csv(loadpath + "test.csv")
train.set_index('PassengerId',inplace=True)
test.set_index('PassengerId',inplace=True)

labels = train.Survived
# labels = pd.get_dummies(labels,prefix='Survived')



train.drop('Survived', axis=1, inplace=True)

train = preproceed(train)
test = preproceed(test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils import np_utils
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
def baseline_model():
    model = Sequential()
    model.add(Dense(64, activation='relu',input_dim=train.shape[1]))
    model.add(keras.layers.core.Dropout(0.2))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(2,activation='softmax'))

    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=baseline_model, epochs=100)
```

```python
model.fit(train, labels, batch_size=100, verbose=True)
```

```python
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=2)
scores = cross_validate(model, X=train, y=labels, cv=kf, scoring='accuracy')
```

```python
scores['test_score']
```

```python
pred = model.predict(test)
```

```python
# create submission file
submission_result = pd.DataFrame({"Survived":pred},index=test.index)
submission_result.to_csv(savepath + 'submission_dnn.csv',index='False')
```
