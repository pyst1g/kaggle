```python
import matplotlib.pyplot as plt
import numpy as np
import optuna
import pandas as pd
import pydotplus
import seaborn as sns

from functools import partial
from graphviz import Digraph
from IPython.display import Image
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.externals.six import StringIO
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc, accuracy_score
from sklearn.model_selection import cross_validate, train_test_split, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier

%matplotlib inline
```

```python
#csvデータの保存場所
savepath = "../input/"
```

* csv呼び出し

```python
df = pd.read_csv(savepath + "train.csv")
df_test = pd.read_csv(savepath + "test.csv")
sampleSubmission = pd.read_csv(savepath + "sampleSubmission.csv")

df.set_index('id',inplace=True)
df_test.set_index('id',inplace=True)
```

```python
train_X = df.drop('target', axis=1)
train_y = df.target
test_X = df_test

clf = DecisionTreeClassifier(min_samples_leaf=4,random_state=2)
clf = clf.fit(train_X, train_y)
test_y = clf.predict_proba(test_X)
```

```python
features = train_X.columns
importances = clf.feature_importances_
indices = np.argsort(importances)

plt.figure(figsize=(6,6))
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), features[indices])
plt.show()
```

```python
def objective(X, y, trial):
    params = {
        'n_estimators': int(trial.suggest_int('n_estimators',1e+2,1e+3)),
        'max_depth': int(trial.suggest_int('max_depth', 2, 32))
    }
    model = RandomForestClassifier(random_state=2,**params)
    
    kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=2)
    scores = cross_validate(model, X=X, y=y, cv=kf, scoring='neg_log_loss')
    
    return 1.0 - scores['test_score'].mean()
```

```python
f = partial(objective, train_X, train_y)
study = optuna.create_study()
study.optimize(f, n_trials=10)
```

```python
study.best_params
```

```python
clf = RandomForestClassifier(random_state=2,**study.best_params).fit(train_X,train_y)
test_y = clf.predict_proba(test_X)
```

```python
submission_result = pd.DataFrame(test_y, index=df_test.index, columns=['Class_1','Class_2','Class_3','Class_4','Class_5','Class_6','Class_7','Class_8','Class_9'])
```

```python
submission_result.to_csv(savepath + 'submission.csv',index=True)
```
